```python
from datamule import PremiumDownloader, Downloader,Portfolio

# Download all 10-Ks for Microsoft
downloader = Downloader()

downloader.download_submissions(submission_type=['10-K'],ticker='MSFT',output_dir='ingest/msft10k',filing_date=('2001-01-01','2025-01-01'))
```

```python
# Load the 10-Ks into a portfolio.
portfolio = Portfolio('ingest/msft10k')

mda_list = []
# iterate over each submission in the portfolio
for submission in portfolio:
    # get all document types that are 10-Ks (other options might be EX-22.1, GRAPHIC, etc.)
    for tenk in submission.document_type(['10-K']):
        # Parse the 10-K
        tenk.parse()
        mda_list.append(tenk.data['document']['part2']['item7'])

print(mda_list[0]) # Print the first 10-K's MD&A section
```
