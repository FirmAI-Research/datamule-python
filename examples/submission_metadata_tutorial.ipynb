## Differences between variants

Both EDGAR and archive SGML contain the same information. However, there are differences in key value storage. Archive variant uses dashes, while EDGAR variant uses spaces. Additionally, some values are abbreviated depending on the variant.

## Datamule's Submission Structure

Datamule parses the SGML submission file and splits them into the individual documents. It also creates a metadata.json file which contains the parsed metadata.

For example, the downloaded file structure for Tesla's 2022 10-K would look like

000095017022000796/

Metadata:
* metadata.json

Primary Documents:
* tsla-20211231.htm (10-K)
* tsla-ex21_1.htm (EX-21.1)
* tsla-ex23_1.htm (EX-23.1)  
* tsla-ex31_1.htm (EX-31.1)
* tsla-ex31_2.htm (EX-31.2)
* tsla-ex32_1.htm (EX-32.1)
* img96779317_0.jpg (GRAPHIC)
...

Data Files:
* tsla-20211231_def.xml (EX-101.DEF)
* tsla-20211231_pre.xml (EX-101.PRE)
* tsla-20211231_lab.xml (EX-101.LAB)
* tsla-20211231.xsd (EX-101.SCH)
* tsla-20211231_cal.xml (EX-101.CAL)
...

We'll now look at the metadata.json file for tesla

````python
from datamule import PremiumDownloader, Downloader

downloader = PremiumDownloader() # Downloader also works, but PremiumDownloader is faster
downloader.download_submissions(submission_type=['10-K'],ticker = 'TSLA', filing_date = ('2022-01-01', '2022-12-31'), output_dir = 'ingest/tesla')
````

## Interacting with metadata using datamule

The easiest way to use metadata in datamule is through the Portfolio class.

````python
from datamule import Portfolio

# multithreaded to be faster
portfolio = Portfolio('ingest/tesla')

for submission in portfolio:

    # accessing a specific key in the metadata
    print(submission.metadata['submission']['ACCESSION-NUMBER'])
````

## Example

Simple Demo for creating a datasets of attachments to 10-K with metadata. Requested via email from a user who wanted an alternative to a product offered by WRDS SEC Suites. If you're reading this, hi :)

The demo will focus on Apple, and will create a csv with:

* filename - the filename of a file in the submission, e.g. a10-kexhibit10202017.htm
* submission_type - the type of submission, e.g. 10-K
* document_type - the type of document, e.g. GRAPHIC, EX-22.1, 10-K. (e.g. the form and its attachments)
* filing_date
* report_date - reporting period, e.g. 20170930
* report_year - extracted from first 4 digits of reporting period
* cik - company's unique identifier
* ticker - company's current ticker - see cautionary note
* company_conformed_name - company's name, e.g. APPLE INC

````python
# Download the data
from datamule import Downloader

downloader = Downloader()

# Download the data
downloader.download_submissions(submission_type=['10-K'],ticker='AAPL',output_dir='ingest/appl_10k')
````

````python
from datamule import Portfolio, load_package_dataset
import pandas as pd

portfolio = Portfolio('ingest/appl_10k')

# Create lists to store document data
data = []

# Note: You may have to adjust the metadata keys to match the structure of the data
for submission in portfolio:
    metadata = submission.metadata
    sub_info = metadata['submission']
    
    # Extract base submission info
    base_info = {
        'submission_type': sub_info['CONFORMED SUBMISSION TYPE'],
        'filing_date': sub_info['FILED AS OF DATE'],
        'report_date': sub_info['CONFORMED PERIOD OF REPORT'],
        'report_year': sub_info['CONFORMED PERIOD OF REPORT'][:4],
        'cik': sub_info['FILER']['COMPANY DATA']['CENTRAL INDEX KEY'],
        'company_conformed_name': sub_info['FILER']['COMPANY DATA']['COMPANY CONFORMED NAME'],
    }
    
    # Add each document with the base info
    for doc in metadata['documents']:  # documents is at root level
        row = base_info.copy()
        row['filename'] = doc['FILENAME']
        row['document_type'] = doc['TYPE']
        data.append(row)

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df = df[['filename', 'submission_type', 'document_type', 'filing_date', 
         'report_date', 'report_year', 'cik', 'company_conformed_name']]

# now we merge in ticker data
ticker_df = pd.DataFrame(load_package_dataset('company_tickers')) # dataset is stored as a list of dictionaries so as to allow alternatives to pandas, like polars.
ticker_df = ticker_df[['cik', 'ticker']]
df = df.merge(ticker_df, on='cik', how='left')

df.to_csv('data/apple_10k_attachments.csv', index=False)
````

Cautionary note: Tickers change over time. `company_tickers` dataset is based on the current SEC provided ticker dataset.
